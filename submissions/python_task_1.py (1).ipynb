```python
import pandas as pd

def generate_car_matrix(df):
    result_df = df.pivot_table(values='car', index='id_1', columns='id_2', fill_value=0)
    result_df.values[[range(len(result_df))]*2] = 0
    return result_df
df = pd.read_csv('/content/drive/My Drive/dataset-1.csv')
result_dataframe = generate_car_matrix(df)
print(result_dataframe)
```

```python
import pandas as pd

def get_type_count(df):

    conditions = [
        (df['car'] <= 15),
        (df['car'] > 15) & (df['car'] <= 25),
        (df['car'] > 25)
    ]


    car_type_values = ['low', 'medium', 'high']
    df['car_type'] = pd.cut(df['car'], bins=[float('-inf'), 15, 25, float('inf')], labels=car_type_values)


    type_count = df['car_type'].value_counts().to_dict()

    sorted_type_count = dict(sorted(type_count.items()))

    return sorted_type_count
result_type_count = get_type_count(df)
print(result_type_count)
```

```python
import pandas as pd

def get_bus_indexes(df):
    # Calculate the mean value of the 'bus' column
    mean_bus_value = df['bus'].mean()

    # Identify indices where bus values are greater than twice the mean
    bus_indexes = df[df['bus'] > 2 * mean_bus_value].index.tolist()

    # Sort the indices in ascending order
    bus_indexes.sort()

    return bus_indexes

# Example usage:
# Assuming df is your DataFrame loaded from dataset-1.csv
# df = pd.read_csv('dataset-1.csv')

# Call the get_bus_indexes function
result_bus_indexes = get_bus_indexes(df)

# Print the result
print(result_bus_indexes)

```

QUESTION 1 (TASK 4)

```python
import pandas as pd

def filter_routes(df):
    # Group by 'route' and calculate the average of 'truck' column for each group
    route_avg_truck = df.groupby('route')['truck'].mean()

    # Filter routes where the average of 'truck' column is greater than 7
    selected_routes = route_avg_truck[route_avg_truck > 7].index.tolist()

    # Sort the selected routes in ascending order
    selected_routes.sort()

    return selected_routes

# Example usage:
# Assuming df is your DataFrame loaded from dataset-1.csv
# df = pd.read_csv('dataset-1.csv')

# Call the filter_routes function
result_selected_routes = filter_routes(df)

# Print the result
print(result_selected_routes)
```

```python
import pandas as pd

def multiply_matrix(result_df):
    # Create a copy of the input DataFrame to avoid modifying the original
    modified_df = result_df.copy()

    # Apply the specified logic to modify values in the DataFrame
    modified_df[modified_df > 20] *= 0.75
    modified_df[modified_df <= 20] *= 1.25

    # Round the modified values to 1 decimal place
    modified_df = modified_df.round(1)

    return modified_df

# Example usage:
# Assuming result_df is the DataFrame obtained from Question 1
result_df = generate_car_matrix(df)

# Call the multiply_matrix function
modified_result_df = multiply_matrix(result_df)

# Print the modified result
print(modified_result_df)
```

```python
import pandas as pd

def check_completeness(df):
    try:
        # Convert 'timestamp' columns to datetime type
        df['start_datetime'] = pd.to_datetime(df['startDay'] + ' ' + df['startTime'])
        df['end_datetime'] = pd.to_datetime(df['endDay'] + ' ' + df['endTime'])

    except pd.errors.OutOfBoundsDatetime as e:
        print(f"Error: {e}")
        return None

    # Calculate the time duration for each row
    df['duration'] = df['end_datetime'] - df['start_datetime']

    # Check if each time duration covers a full 24-hour period and spans all 7 days
    completeness_check = (
        (df['duration'] == pd.Timedelta(days=1)) &  # Covers 24 hours
        (df['start_datetime'].dt.day_name() == df['end_datetime'].dt.day_name())  # Spans the same day of the week
    )

    # Group by (id, id_2) and check if all rows for each group pass the completeness check
    result_series = completeness_check.groupby(['id', 'id_2']).all()

    return result_series

# Example usage:
# Assuming df is your DataFrame loaded from dataset-2.csv
# df = pd.read_csv('dataset-2.csv')

# Call the check_completeness function
completeness_result = check_completeness(df)

# Print the result
print(completeness_result)
```
