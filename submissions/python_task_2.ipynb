```python
import pandas as pd

def calculate_distance_matrix(file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv('/content/drive/My Drive/dataset-3.csv')

    # Create a list to store unique toll locations
    toll_locations = list(set(df['id_start'].unique()) | set(df['id_end'].unique()))

    # Create an empty DataFrame to store the distance matrix
    distance_matrix = pd.DataFrame(index=toll_locations, columns=toll_locations)

    # Initialize diagonal values to 0
    distance_matrix.values[[range(len(toll_locations))]*2] = 0

    # Calculate distances and fill in the matrix
    for from_location in toll_locations:
        for to_location in toll_locations:
            if from_location == to_location:
                continue  # Skip diagonal elements
            else:
                # Find all rows where the distance is known
                forward_distance = df.loc[(df['id_start'] == from_location) & (df['id_end'] == to_location), 'distance'].sum()
                backward_distance = df.loc[(df['id_start'] == to_location) & (df['id_end'] == from_location), 'distance'].sum()

                # Calculate the cumulative distance along known routes
                cumulative_distance = forward_distance + backward_distance

                # Fill in the matrix symmetrically
                distance_matrix.loc[from_location, to_location] = cumulative_distance
                distance_matrix.loc[to_location, from_location] = cumulative_distance

    return distance_matrix

file_path = 'dataset-3.csv'
result_matrix = calculate_distance_matrix(file_path)
print(result_matrix)
```

```python
import pandas as pd

def unroll_distance_matrix(distance_matrix):
    # Create an empty DataFrame to store unrolled distance values
    unrolled_df = pd.DataFrame(columns=['id_start', 'id_end', 'distance'])

    # Iterate through the distance matrix
    for id_start in distance_matrix.index:
        for id_end in distance_matrix.columns:
            if id_start != id_end:
                # Extract distance value from the distance matrix
                distance = distance_matrix.loc[id_start, id_end]

                # Append the data to the unrolled DataFrame
                unrolled_df = unrolled_df.append({'id_start': id_start, 'id_end': id_end, 'distance': distance}, ignore_index=True)

    return unrolled_df

# Example usage:
file_path = 'dataset-3.csv'
distance_matrix = calculate_distance_matrix(file_path)  # Assuming you have the function from Question 1
unrolled_df = unroll_distance_matrix(distance_matrix)
print(unrolled_df)
```

```python
import pandas as pd

def find_ids_within_ten_percentage_threshold(df, reference_value):
    # Filter rows for the given reference value
    reference_rows = df[df['id_start'] == reference_value]

    if reference_rows.empty:
        raise ValueError(f"No data found for reference value {reference_value}")

    # Calculate the average distance for the reference value
    average_distance = reference_rows['distance'].mean()

    # Calculate the threshold range (10% of the average distance)
    threshold_lower = 0.9 * average_distance
    threshold_upper = 1.1 * average_distance

    print(f"Reference Value: {reference_value}")
    print(f"Average Distance: {average_distance}")
    print(f"Threshold Range: {threshold_lower} to {threshold_upper}")

    # Filter rows within the threshold range
    within_threshold_rows = df[(df['distance'] >= threshold_lower) & (df['distance'] <= threshold_upper)]

    # Get unique values from 'id_start' column and sort them
    result_ids = sorted(within_threshold_rows['id_start'].unique())

    return result_ids

# Example usage:
file_path = 'dataset-3.csv'
distance_matrix = calculate_distance_matrix(file_path)  # Assuming you have the distance matrix
unrolled_df = unroll_distance_matrix(distance_matrix)    # Assuming you have the unrolled DataFrame

# Choose a reference value that exists in the 'id_start' column of your DataFrame
reference_value = unrolled_df['id_start'].sample().iloc[0]

result_ids = find_ids_within_ten_percentage_threshold(unrolled_df, reference_value)
print(result_ids)
```

```python
import pandas as pd

def calculate_toll_rate(df):
    # Define rate coefficients for each vehicle type
    rate_coefficients = {'moto': 0.8, 'car': 1.2, 'rv': 1.5, 'bus': 2.2, 'truck': 3.6}

    # Create new columns for each vehicle type
    for vehicle_type, rate_coefficient in rate_coefficients.items():
        df[vehicle_type] = df['distance'] * rate_coefficient

    return df

# Example usage:
file_path = 'dataset-3.csv'
distance_matrix = calculate_distance_matrix(file_path)  # Assuming you have the distance matrix
unrolled_df = unroll_distance_matrix(distance_matrix)    # Assuming you have the unrolled DataFrame

# Add toll rate columns
result_df = calculate_toll_rate(unrolled_df)
print(result_df)
```

```python
import pandas as pd
from datetime import time

def calculate_time_based_toll_rates(df):
    # Define time ranges and discount factors
    time_ranges = [
        (time(0, 0, 0), time(10, 0, 0), 0.8),
        (time(10, 0, 0), time(18, 0, 0), 1.2),
        (time(18, 0, 0), time(23, 59, 59), 0.8)
    ]
    weekend_discount_factor = 0.7

    # Create new columns for time-based toll rates
    for start_time, end_time, discount_factor in time_ranges:
        mask = (df['start_time'] >= start_time) & (df['end_time'] <= end_time) & (df['start_day'] < 'Saturday')
        df.loc[mask, 'weekday_toll_rate'] = df.loc[mask, 'distance'] * discount_factor

    # Apply constant discount factor for weekends
    weekend_mask = (df['start_day'] >= 'Saturday')
    df.loc[weekend_mask, 'weekend_toll_rate'] = df.loc[weekend_mask, 'distance'] * weekend_discount_factor

    return df

# Example usage:
file_path = 'dataset-3.csv'
distance_matrix = calculate_distance_matrix(file_path)  # Assuming you have the distance matrix
unrolled_df = unroll_distance_matrix(distance_matrix)    # Assuming you have the unrolled DataFrame

# Assuming your DataFrame has columns 'start_day', 'start_time', 'end_day', 'end_time', and 'distance'
result_df = calculate_time_based_toll_rates(unrolled_df)
print(result_df)
```
